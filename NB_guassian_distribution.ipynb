```python
import numpy as np
import sklearn
import matplotlib.pyplot as plt
import math
import random
from mpl_toolkits.mplot3d import Axes3D
from scipy.stats import multivariate_normal
```

```python
x = np.linspace(0, 5, 100, endpoint=False)
y = multivariate_normal.pdf(x, mean=2.5, cov=0.5)
fig1 = plt.figure()
ax = fig1.add_subplot(111)
ax.plot(x, y)
plt.show()
```

```python
def mvg_pdf(x,y,mean,cov,D):
    X = np.array([x,y])
    mu = np.array([mean]*D)
    return 1/(2*math.pi**(D/2)*np.sqrt(np.linalg.det(cov)))*np.exp(-1/2 * ((X-mu).T@np.linalg.inv(cov)@(X-mu)))
```

```python
D = 2
n = 100
mean = 2.5
cov = np.array([[2, 1],[1, 2]])
x_linspace = np.linspace(0,5,n)
y_linspace = np.linspace(0,5,n)
z = np.zeros((n,n))
i = 0
j = 0

for x in x_linspace:
    for y in y_linspace:
        z[i,j] = mvg_pdf(x,y,mean,cov,D)
        j += 1
    i += 1
    j = 0
```

```python
X,Y = np.meshgrid(x_linspace,y_linspace)
fig = plt.figure(figsize=(6,6))
ax = fig.add_subplot(111, projection='3d')


# Plot a 3D surface
ax.plot_surface(X, Y, z)

plt.show()
```

```python
X,Y = np.meshgrid(x_linspace,y_linspace)
fig = plt.figure(figsize=(6,6))
fig,ax=plt.subplots(1,1)
cp = ax.contourf(X, Y, z)
fig.colorbar(cp)
plt.show()
```
